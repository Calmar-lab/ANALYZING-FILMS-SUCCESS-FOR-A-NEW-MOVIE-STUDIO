```python
#import python libraries needed
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load  the movie_budgets csv file
df1=pd.read_csv("tn.movie_budgets.csv.gz")
#print the first 10 rows
df1.head(10)
```

```python
# Connect to the SQLite database 
conn = sqlite3.connect("im.db/im.db")

# List all tables in the database
df2= pd.read_sql("""SELECT name FROM sqlite_master;""", conn)
df2
```

```python
#Load the movie_basics table
movie_basics_df=pd.read_sql(""" SELECT * FROM movie_basics ;""", conn)
movie_basics_df
```

```python
#Load the movie_ratings table
movie_ratings_df= pd.read_sql(""" SELECT * FROM movie_ratings ;""", conn)
movie_ratings_df
```

```python
#Load the directors table
directors_df= pd.read_sql(""" SELECT * FROM directors;""", conn)
directors_df
```

```python
#Load the writers table
writers_df=pd.read_sql(""" SELECT * FROM writers;""", conn)
writers_df
```

```python
#Merge the table movie_basics with movie_ratings using the column movie_id
movie_basics_ratings_df=pd.merge(movie_basics_df, movie_ratings_df, on='movie_id', how='left')
movie_basics_ratings_df
```

```python
# Fill missing 'original_title' with 'primary_title' values
movie_basics_ratings_df['original_title'] = movie_basics_ratings_df['original_title'].fillna(movie_basics_ratings_df['primary_title'])

# Fill missing 'runtime_minutes' with the median
movie_basics_ratings_df['runtime_minutes'] = movie_basics_ratings_df['runtime_minutes'].fillna(movie_basics_ratings_df['runtime_minutes'].median())

# Fill missing 'genres' with "Unknown"
movie_basics_ratings_df['genres'] = movie_basics_ratings_df['genres'].fillna("Unknown")

# Fill missing 'averagerating' with the mean
movie_basics_ratings_df['averagerating'] = movie_basics_ratings_df['averagerating'].fillna(movie_basics_ratings_df['averagerating'].mean())

# Fill missing 'numvotes' with the median
movie_basics_ratings_df['numvotes'] = movie_basics_ratings_df['numvotes'].fillna(movie_basics_ratings_df['numvotes'].median())
```

***EXPLOTATORY DATA ANALYSIS***


**1. The average return of films**

```python
# Removing  non-numeric characters and converting them to float
df1['worldwide_gross'] = (df1['worldwide_gross']
    .replace('[\$,]', '', regex=True)
    .astype(float)
)
# Calculating the mean of worldwide gross revenue
mean_worldwide_gross = df1['worldwide_gross'].mean()

# Plotting the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df1['worldwide_gross'], bins=30, kde=True, color='skyblue')
plt.axvline(mean_worldwide_gross, color='red', linestyle='--', label=f'Mean: ${mean_worldwide_gross:.2f}')
plt.title('Distribution of Worldwide Gross Revenue')
plt.xlabel('Worldwide Gross Revenue')
plt.ylabel('Count')
plt.legend()
plt.savefig('Distribution of Worldwide Gross Revenue')
plt.show();
```

```python
"""
The Average movie revenue is $91487460.91 with a Right_Skewed Distribution which implies that most movies earn below averages. 
This can be used by the Company as a reference point when making decisions on the Production costs.
This mean can also be used as a reference point whether their movies perform below or above the market revenue.
"""
```

***2. Number of Movies Produced per Year***

```python
# Ensuring the 'genres' column is converted to strings and also combining any lists into comma-separated strings
filtered_df = movie_basics_ratings_df[
    (movie_basics_ratings_df['start_year'] >= 2010) &
    (movie_basics_ratings_df['start_year'] <= 2024)
].copy()
# Group by 'start_year' and count the number of movies per year
movies_per_year = filtered_df['start_year'].value_counts().sort_index()

# Plot the data
plt.figure(figsize=(12, 6))
sns.lineplot(x=movies_per_year.index, y=movies_per_year.values, marker='o', color='blue')
plt.xlabel('Year')
plt.ylabel('Number of Movies Produced')
plt.title('Number of Movies Produced per Year')
plt.grid(True)
plt.savefig('Number of Movies Produced per Year')
plt.show()
```
