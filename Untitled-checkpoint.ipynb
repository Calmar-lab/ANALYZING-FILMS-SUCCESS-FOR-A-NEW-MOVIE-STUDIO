```python
# Fill missing 'original_title' with 'primary_title' values
movie_basics_ratings_df['original_title'] = movie_basics_ratings_df['original_title'].fillna(movie_basics_ratings_df['primary_title'])

# Fill missing 'runtime_minutes' with the median
movie_basics_ratings_df['runtime_minutes'] = movie_basics_ratings_df['runtime_minutes'].fillna(movie_basics_ratings_df['runtime_minutes'].median())

# Fill missing 'genres' with "Unknown"
movie_basics_ratings_df['genres'] = movie_basics_ratings_df['genres'].fillna("Unknown")

# Fill missing 'averagerating' with the mean
movie_basics_ratings_df['averagerating'] = movie_basics_ratings_df['averagerating'].fillna(movie_basics_ratings_df['averagerating'].mean())

# Fill missing 'numvotes' with the median
movie_basics_ratings_df['numvotes'] = movie_basics_ratings_df['numvotes'].fillna(movie_basics_ratings_df['numvotes'].median())
```

***EXPLOTATORY DATA ANALYSIS***


**1. The average return of films**

```python
# Removing  non-numeric characters and converting them to float
df1['worldwide_gross'] = (df1['worldwide_gross']
    .replace('[\$,]', '', regex=True)
    .astype(float)
)
# Calculating the mean of worldwide gross revenue
mean_worldwide_gross = df1['worldwide_gross'].mean()

# Plotting the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df1['worldwide_gross'], bins=30, kde=True, color='green')
plt.axvline(mean_worldwide_gross, color='red', linestyle='--', label=f'Mean: ${mean_worldwide_gross:.2f}')
plt.title('Distribution of Worldwide Gross Revenue')
plt.xlabel('Worldwide Gross Revenue')
plt.ylabel('Count')
plt.legend()
plt.show();
```

***2. The trends of genre production over the years***

```python
# Filter data for years 2011 to 2024
filtered_df = movie_basics_ratings_df[
    (movie_basics_ratings_df['start_year'] >= 2009) &
    (movie_basics_ratings_df['start_year'] <= 2024)
]

# Group by genre and year, calculating the average rating
genre_performance_df = (
    filtered_df.groupby(['genres', 'start_year'])['averagerating']
    .mean()
    .reset_index()
)

# Select the top 10 most common genres in this filtered range
top_genres = genre_performance_df['genres'].value_counts().nlargest(10).index
filtered_top_genres_df = genre_performance_df[genre_performance_df['genres'].isin(top_genres)]

# Plot line graph for genre performance over years 2011-2024
plt.figure(figsize=(12, 8))
for genre in top_genres:
    genre_data = filtered_top_genres_df[filtered_top_genres_df['genres'] == genre]
    plt.plot(genre_data['start_year'], genre_data['averagerating'], label=genre)

plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Top 10 Genre Performance (Average Rating) from 2011 to 2024')
plt.legend(title='Genre')
plt.grid(True)
plt.show()
```
